```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('Melbourne_housing_FULL.csv')

# Checking initial missing values
initial_missing = df.isnull().sum()

# Handling missing values
# Filling missing numeric values with the median
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.median()), axis=0)

# Filling missing categorical values with the mode
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]), axis=0)

# Normalizing numeric columns
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Checking final missing values
final_missing = df.isnull().sum()

# Plotting the initial and final missing values
fig, ax = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
initial_missing.plot(kind='bar', ax=ax[0], title='Initial Missing Values', color='red')
final_missing.plot(kind='bar', ax=ax[1], title='Final Missing Values', color='green')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Melbourne_housing_FULL.csv')

# Sample a smaller part of the dataset if it's very large
df_sampled = df.dropna(subset=['Lattitude', 'Longtitude']).sample(frac=0.05, random_state=42)

# Selecting the features for clustering
coordinates_sampled = df_sampled[['Lattitude', 'Longtitude']]

# Scaling the features for clustering
scaler = StandardScaler()
coordinates_scaled = scaler.fit_transform(coordinates_sampled)

# Using the elbow method to find the optimal number of clusters
wcss_sampled = []  # Within-cluster sums of squares for the sampled data
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(coordinates_scaled)
    wcss_sampled.append(kmeans.inertia_)

# Creating the elbow plot
plt.figure(figsize=(10, 8))
plt.plot(range(1, 11), wcss_sampled, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (within-cluster sums of squares)')
plt.grid(True)
plt.show()

# From the elbow plot, determine the optimal number of clusters and set it here
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(coordinates_scaled)

# Add the clustering results back to the original dataframe
df.loc[df_sampled.index, 'Cluster'] = clusters

# The following plotting code should not be indented if it's not part of a larger code block
# Plotting code for the clustering plot (to be executed separately to visualize the results)
plt.figure(figsize=(10, 8))
plt.scatter(df_sampled['Longtitude'], df_sampled['Lattitude'], c=clusters, cmap='viridis', s=50)
plt.title('K-Means Clustering of Properties by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```
